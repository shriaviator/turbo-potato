```python
import pandas as pd 
import altair as alt 
```

```python
pd.set_option('display.max_rows', 85)
```

```python
df = pd.read_csv('disc_instructor.csv')
df
```

```python

def mask_list(df,maskname):
  tempdf = df.copy(deep=True)
  final_list=[]
  for loc,xray in enumerate(tempdf['name'].tolist()):
    if xray in maskname:
      final_list.append(xray)
    else:
      final_list.append(loc)
  
  tempdf['name'] = final_list
  mask = tempdf.name.apply(lambda x: any(item for item in maskname if item == x))
  tempdf['status'] = tempdf[mask]
  return tempdf
    
```

```python
maskname = ['Dr. Malolan Sundararaman','Nikita Kumari','Karthik POD']
tempdf = df.copy(deep=True)
mask = tempdf.name.apply(lambda x: any(item for item in maskname if item == x))
tempdf['status'] = mask
tempdf

domain = [True, False ]
range_ = ['red', 'green']

# req_df = mask_list(tempdf,['Dr. Malolan Sundararaman','Nikita Kumari','Karthik POD'])
scatter = alt.Chart(tempdf).mark_point().encode(
    x=alt.X('name:N',sort='-y', axis=alt.Axis( title='Name of Instructor'),labelColor=alt.condition(X in maskname , alt.value('red'), alt.value('black'))),
    y=alt.Y('visits',axis=alt.Axis( title='Number of Days Visited')),
    
    color=alt.Color('status:N', scale=alt.Scale(domain=domain, range=range_),legend=None)
    ,
    shape = alt.Shape('status',legend=None),
    tooltip=['name','visits']

)
rule = alt.Chart(tempdf).mark_rule(color='red').encode(
    y='mean(visits):Q'
)

(scatter + rule).properties(width=1200).interactive()
```
